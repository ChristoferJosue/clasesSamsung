## ejemplo 

```python
# con memoización
storage = dict([
    (0,1),
    (1,1)
]) 
def fib(n):
  if n in storage:
    return storage[n]
  storage[n] = fib(n-1) + fib(n-2)
  return storage[n]
```

```python
%%time
fib(35)
```

```python
# sin memoización
def fib(n):
  if n == 0 or n == 1:
    return 1
  return fib(n-1) + fib(n-2)
 
```

```python
%%time
fib(35)
```
